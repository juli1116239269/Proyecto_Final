#Cargar bases de datos, asignar semestres, grupos y concatenar

```python
# Generar listas apartir de los nombres y apeelidos, y juntarlos en una sola
def GenerarNombre(Nombres: list, Apellidos: list) -> str:
    Nombre = rnd.choice(Nombres)
    Apellido = rnd.choice(Apellidos)
    return f'{Nombre.upper()}_{Apellido.upper()}'
```

```python
#Leer los archivos que se subieron a GitHub y en caso de que haya espacion entre los nombres o apellidos de una persona va poner guion bajo

dfNombres = pd.read_csv("https://github.com/juli1116239269/TrabajoFinal_prueba/raw/59e5f80239484d00480e37270bc1529035f5847c/NombresArgentina.csv", encoding="latin1")
Nombres = dfNombres['name'].tolist()
for i in tqdm.trange(len(Nombres)):
    if ' ' in Nombres[i]:
        Nombres[i]=Nombres[i].replace(' ', '_')

dfApellidos = pd.read_csv("https://github.com/juli1116239269/TrabajoFinal_prueba/raw/main/ApellidosArgentina.csv", encoding="latin1")
Apellidos = dfApellidos['lastname'].tolist()
for i in tqdm.trange(len(Apellidos)):
    if ' ' in Apellidos[i]:
        Apellidos[i]=Apellidos[i].replace(' ', '_')

df1 = pd.DataFrame(columns=['Nombre&Apellido'])
for i in tqdm.trange(1000):
    vector = []
    nombre = GenerarNombre(Nombres, Apellidos)
    vector = [nombre]
    df1.loc[len(df1)] = vector
```

```python
# Crear semestres
def GenerearSemestre() -> int:
    semestre = [1] * 140
    semestre.extend([2] * 130)
    semestre.extend([3] * 120)
    semestre.extend([4] * 110)
    semestre.extend([5] * 100)
    semestre.extend([6] * 100)
    semestre.extend([7] * 90)
    semestre.extend([8] * 80)
    semestre.extend([9] * 70)
    semestre.extend([10] * 60)

    rnd.shuffle(semestre)
    return semestre

semestres_generados = GenerearSemestre()
df2 = pd.DataFrame({'SEMESTRE': semestres_generados})
```

```python
# Concatenar nombre y apellido con el semestre y definir a que grupo para cada asignatura va pertenecer la persona, teniendo en cuentas la cantidad de grupos para cada semestre y su capacidad maxima
personas = pd.concat([df1, df2], axis=1)

def grupos(SEMESTRE):
    if SEMESTRE == 1:
        return np.repeat([1, 2, 3, 4, 5], [30, 30, 30, 30, 20])
    elif SEMESTRE == 2:
        return np.repeat([1, 2, 3, 4, 5], [30, 30, 30, 30, 10])
    elif SEMESTRE == 3:
        return np.repeat([1, 2, 3, 4], [30, 30, 30, 30])
    elif SEMESTRE == 4:
        return np.repeat([1, 2, 3, 4, 5], [25, 25, 25, 25, 10])
    elif SEMESTRE == 5:
        return np.repeat([1, 2, 3, 4], [25, 25, 25, 25])
    elif SEMESTRE == 6:
        return np.repeat([1, 2, 3, 4], [25, 25, 25, 25])
    elif SEMESTRE == 7:
        return np.repeat([1, 2, 3, 4, 5], [20, 20, 20, 20, 20])
    elif SEMESTRE == 8:
        return np.repeat([1, 2, 3, 4], [20, 20, 20, 20])
    elif SEMESTRE == 9:
        return np.repeat([1, 2, 3, 4], [20, 20, 20, 20])
    else:
        return np.repeat([1, 2, 3, 4, 5, 6], [10, 10, 10, 10, 10, 10])

# Crear una columna de grupos vacía
personas['GRUPO'] = 0

# Asignar los grupos para cada semestre
for sem in personas['SEMESTRE'].unique():
    mask = personas['SEMESTRE'] == sem
    grupos_asignados = grupos(sem)
    personas.loc[mask, 'GRUPO'] = grupos_asignados[:len(personas[mask])]  # Asegurarse de asignar el número correcto de grupos


personas
```

#Generar código para cada asignatura (CA)

```python
def generar_numero_aleatorio(unicos):
    while True:
        numero = rnd.randint(0, 9)
        if numero not in unicos:
            unicos.add(numero)
            return numero


def generar_codigo(materia, codigos_existentes, unicos):
    # Tomar las iniciales de la materia (primeros 3 caracteres)
    iniciales_materia = materia['ASIGNATURA'][:3].upper()
    # Tomar el semestre (primer dígito de la columna 'semestre')
    semestre = str(materia['SEMESTRE'])

    # Generar un número aleatorio único
    while True:
        numero_aleatorio = str(rnd.randint(0, 9))
        CA = iniciales_materia + semestre + numero_aleatorio
        if CA not in codigos_existentes:
            codigos_existentes.add(CA)
            return CA

# Conjunto para almacenar los códigos únicos generados
codigos_existentes = set()

asignaturas['CA'] = asignaturas.apply(lambda materia: generar_codigo(materia, codigos_existentes, set()), axis=1)
asignaturas
```

```python
logging.info("Se generaron los codigos para cada asignatura")
```

#Generar número de aulas para cada asignatura (TCA)

```python

def generar_aulas(SEMESTRE):
    if SEMESTRE == 1:
      return 140/30
    elif SEMESTRE == 2:
      return  130/30
    elif SEMESTRE == 3:
      return 120/30
    elif SEMESTRE == 4:
      return 110/25
    elif SEMESTRE == 5:
      return 100/25
    elif SEMESTRE == 6:
      return 100/25
    elif SEMESTRE == 7:
      return 90/20
    elif SEMESTRE == 8:
      return 80/20
    elif SEMESTRE == 9:
      return 70/20
    else:
      return 60/10

asignaturas['TCA'] = asignaturas['SEMESTRE'].apply(generar_aulas)
asignaturas["TCA"] = asignaturas["TCA"].apply(math.ceil)
asignaturas
```

# Generar capacidad por aula para cada asignatura (CAP)

```python
def generar_capacidad_aula(SEMESTRE):
    if SEMESTRE == 1:
      return 30
    elif SEMESTRE == 2:
      return  30
    elif SEMESTRE == 3:
      return 30
    elif SEMESTRE == 4:
      return 25
    elif SEMESTRE == 5:
      return 25
    elif SEMESTRE == 6:
      return 25
    elif SEMESTRE == 7:
      return 20
    elif SEMESTRE == 8:
      return 20
    elif SEMESTRE == 9:
      return 20
    else:
      return 10

asignaturas['CAP-MAX'] = asignaturas['SEMESTRE'].apply(generar_capacidad_aula)
asignaturas
```

```python
logging.info("Se generaron el numero de aulas necesarias para cada asignatura")
```

# Generar horas de trabajo del docente (HTD)

```python
#  Horas de trabajo docente (HTD)
def calcular_HTD(CREDITOS):
    if CREDITOS == 4:
        return 96
    elif CREDITOS == 3:
        return 64
    elif CREDITOS == 2:
        return 32
    else:
        return 16
asignaturas['HTD'] = asignaturas['CREDITOS'].apply(calcular_HTD)
asignaturas
```

```python
logging.info("Se generaron los horas de trabajo del docente poara cada asigantura")
```

#Generar horas de trabajo independiente (HTI)

```python
# Horas de trabajo independiente (HTI)
def calcular_HTI(CREDITOS):
    if CREDITOS == 4:
        return 120
    elif CREDITOS == 3:
        return 80
    elif CREDITOS == 2:
        return 64
    else:
        return 32
asignaturas['HTI'] = asignaturas['CREDITOS'].apply(calcular_HTI)
asignaturas
```

# Creacion de carpetas para la ruta final del trabajo

```python
# Realizar el merge para combinar los dataframes en función del semestre
ESTUDIANTES = pd.merge(personas, asignaturas, on='SEMESTRE')

# Mostrar el dataframe resultante
ESTUDIANTES = ESTUDIANTES.drop(columns=['CAP-MAX'])
ESTUDIANTES = ESTUDIANTES.drop(columns=['NTE'])
ESTUDIANTES = ESTUDIANTES.drop(columns=['TCA'])
ESTUDIANTES
```

```python
base_path = 'Ruta Trabajo Final'

# Crear las carpetas para cada semestre y asignatura y guardar los archivos Excel
for semestre in asignaturas['SEMESTRE'].unique():
    semestre_path = os.path.join(base_path, f'semestre {semestre}')
    os.makedirs(semestre_path, exist_ok=True)
    for asignatura in asignaturas[asignaturas['SEMESTRE'] == semestre]['ASIGNATURA']:
        asignatura_path = os.path.join(semestre_path, asignatura)
        os.makedirs(asignatura_path, exist_ok=True)

        # Filtrar las personas correspondientes a este semestre
        personas = ESTUDIANTES[ESTUDIANTES['SEMESTRE'] == semestre]
        personas_semestre = personas[personas["ASIGNATURA"] == asignatura]

        # Guardar el DataFrame de personas como un archivo Excel en la carpeta de la asignatura
        archivo_excel = os.path.join(asignatura_path, 'Estudiantes.xlsx')
        personas_semestre.to_excel(archivo_excel, index=False)

        # Guardar el DataFrame de personas como un archivo CSV en la carpeta de la asignatura
        archivo_csv = os.path.join(asignatura_path, 'Estudiantes.csv')
        personas_semestre.to_csv(archivo_csv, index=False)

# Verificar la estructura de carpetas creada
!pip install tree
!tree carpeta
```
